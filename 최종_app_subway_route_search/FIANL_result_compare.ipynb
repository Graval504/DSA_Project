```python
from L_BFS import*
from L_a_star import*
from L_a_star_yen import*
from L_dijkstra import*
from L_bellman_ford import*

#data loading
import json 
data_file ='subway(real).json '

with open(data_file, 'r', encoding='utf-8') as f:
    graph = json.load(f)

#출발지 & 도착지 입력
start = input('출발지: ')
end = input('도착지: ')

#결과 저장
result_bfs=bfs(graph,start,end)
result_dijkstra=dijkstra(graph,start,end)
result_belmanford=bellman_ford(graph,start,end)

''' BFS'''
#후처리
print(start+'  -->  '+end)
print()
print()
print("<알고리즘 이름 : BFS>")
print('출발노선/역: %s호선, %s ' %(result_bfs[0][1],result_bfs[0][0]))
print('도착노선/역: %s호선, %s ' %(result_bfs[-1][1],result_bfs[-1][0]))
# 소요시간 계산 및 환승 처리
print('경로 : ',end='')
# print("(%s %s)"%(result_bfs[0][0],result_bfs[0][1][-1]), end=' ')
leadTime=0
current_line = result_bfs[0][1]
for index, data in  enumerate(result_bfs):
    #if data[0] == start: continue
    if current_line != data[1] : print("-[%s -> %s호선 환승]- "%(current_line,data[1]),end='') # 환승처리
    print("(%s %s)"%(data[0],data[1][-1]), end=' ') #역 및 호선 출력
    current_line = data[1]
    leadTime+=data[2]
print()
print('실제 소요시간(분)',leadTime)
print('총 역의 개수(출발역, 경유역, 도착역 포함): %d개'%len(result_bfs))
print()

''' 다익스트라'''
print("<알고리즘 이름 : 다익스트라>")
print('출발노선/역: %s호선, %s ' %(result_dijkstra[1][0][1],result_dijkstra[1][0][0]))
print('도착노선/역: %s호선, %s ' %(result_dijkstra[1][-1][1],result_dijkstra[1][-1][0]))
# 소요시간 계산 및 환승 처리
print('경로 : ',end='')
# print("#(%s %s)"%(result_dijkstra[1][0][0],result_dijkstra[1][0][1][-1]), end=' ') #출발역

current_line = result_dijkstra[1][0][1]
for index, data in  enumerate(result_dijkstra[1]):
    # if data[0] == start: continue
    if current_line != data[1] : print("-[%s -> %s호선 환승]- "%(current_line,data[1]),end='') # 환승처리
    print("(%s %s)"%(data[0],data[1][-1]), end=' ') #역 및 호선 출력
    current_line = data[1]
print()
print('실제 소요시간(분)',result_dijkstra[0])
print('총 역의 개수(출발역, 경유역, 도착역 포함): %d개'%(len(result_dijkstra[1])))
print()

'''벨만-포드'''
print("<알고리즘 이름 : 벨만-포드>")
print('출발노선/역: %s호선, %s ' %(result_belmanford[1][0][1],result_belmanford[1][0][0]))
print('도착노선/역: %s호선, %s ' %(result_belmanford[1][-1][1],result_belmanford[1][-1][0]))
# 소요시간 계산 및 환승 처리
print('경로 : ',end='')
# print("#(%s %s)"%(result_belmanford[1][0][0],result_belmanford[1][0][1][-1]), end=' ') #출발역

current_line = result_belmanford[1][0][1]
for index, data in  enumerate(result_belmanford[1]):
    # if data[0] == start: continue
    if current_line != data[1] : print("-[%s -> %s호선 환승]- "%(current_line,data[1]),end='') # 환승처리
    print("(%s %s)"%(data[0],data[1][-1]), end=' ') #역 및 호선 출력
    current_line = data[1]
print()
print('실제 소요시간(분)',result_belmanford[0])
print('총 역의 개수(출발역, 경유역, 도착역 포함): %d개'%(len(result_belmanford[1])))
print()
```
