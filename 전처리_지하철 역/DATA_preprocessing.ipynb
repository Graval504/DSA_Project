역간거리 전처리

```python
import pandas as pd
import openpyxl

sheetName = '5호선_마천행'
outputName='5호선-마천행_역간거리.csv'
# 엑셀 파일 불러오기
wb = openpyxl.load_workbook(filename='지하철 역 및 역간 소요시간 수정.xlsx')

# 시트 선택
sheet = wb[sheetName]

# 데이터프레임 생성
data_frame = pd.DataFrame(sheet.values)

data_frame = data_frame.drop(0) # 첫행 제거
data_frame[2]= data_frame[2].astype(str)
# 데이터프레임 출력
print(data_frame)
```

```python
# 2번 열의 각 행 값 읽어오기
values = data_frame.iloc[:, 2]
# print(values)
# print(type(values))

interval =values.tolist()
# print(interval)
interval_new=[]

for i in range(len(interval)):
    min, sec,_=interval[i].split(':')
    weight = int(min)*60+int(sec)
    # print(weight)
    interval_new.append(weight)

print(interval_new)
```

```python
subwayStation=data_frame
print(subwayStation)
subwayStation[2]=interval_new
data_frame
```

```python
subwayStation.to_csv(outputName, header=False,index=False,encoding='cp949')
```
